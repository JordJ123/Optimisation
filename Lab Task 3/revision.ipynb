```python
from scipy.stats import norm
import math
import matplotlib.pyplot as pyplot
import numpy
```

# Task 1

```python
mean = 163
sd = 10
h_rvs = norm.rvs(loc=mean, scale=sd, size=10000) 
print(h_rvs)
```

# Task 2

```python
pyplot.hist(h_rvs)
pyplot.xlabel("height")
pyplot.ylabel("frequency")
pyplot.show()
```

# Task 3

```python
def p(x, sd, mean):
    p1 = 1 / (sd * math.sqrt(2 * math.pi))
    p2 = numpy.exp((-0.5 * ((x - mean) / sd))**2)
    return p1 * p2
```

```python
x = numpy.linspace(mean - 3 * sd, mean + 3 * sd, 1000)
y = p(x, sd, mean)
pyplot.plot(x, y)
pyplot.xlabel("height")
pyplot.ylabel("frequency")
pyplot.show()
```
