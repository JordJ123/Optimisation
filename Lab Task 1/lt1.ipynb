# Task 1

```python
def even(x):
    return (x % 2) == 0
print(even(5))
print(even(122))
```

```python
def percentage(p, m):
    coursework = (p[0]/100) * m[0]
    quiz = (p[1]/100) * m[1]
    return coursework + quiz
p = [65, 35]
m = [35, 21]
print (percentage(p, m))
```

```python
def maximum(a):
    max = a[0]
    for i in a:
        if (i > max):
            max = i
    return max
print(maximum([1,19,2,3,4,100]))
```

```python
def random(m, n):
    return numpy.random.randint(100, 
        size=(m,n))
print(random(3,3))
```

```python
def fib(n):
    if (n == 0):
        return 0
    elif (n == 1):
        return 1
    elif (n > 1):
        return fib(n - 1) + fib(n - 2)
    else:
        return -1
print(fib(20))
```
