```python
import math
import matplotlib.pyplot as pyplot
import numpy
```

# Question 1

```python
def f(x):
    first = pow((6*x - 2), 2)
    second = math.sin(12*x - 4)
    return first * second
```

```python
def g(x):
    return math.sin(10*x)
```

```python
def s(x):
    return 20 * max(0, g(x))
```

```python
def oDeath(x):
    if (g(x) <= 0):
        return f(x)
    else:
        return 20
```

```python
def oStatic(x):
    return f(x) + s(x)
```

# Question 2

```python
x = numpy.linspace(0, 1, 1000)

y = []
for i in x:
    y.append(f(i))
pyplot.plot(x, y)

y = []
for i in x:
    y.append(g(i))
pyplot.plot(x, y)

y = []
for i in x:
    y.append(s(i))
pyplot.plot(x, y)

y = []
for i in x:
    y.append(oDeath(i))
pyplot.plot(x, y)

y = []
for i in x:
    y.append(oStatic(i))
pyplot.plot(x, y)

pyplot.axhline(0, color="black", ls="dotted")
pyplot.xlabel("x")
pyplot.ylabel("y")
pyplot.show()
```

# Question 3

```python
x = numpy.linspace(0, 1, 1000)
y = []
for i in x:
    y.append(f(i))
min = numpy.min(y)

print(x[y.index(min)])
print(min)
```
