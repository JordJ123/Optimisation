```python
import math
import matplotlib.pyplot as pyplot
import numpy
```

# Task 1

```python
def f(x):
    p1 = ((6*x) - 2)**2
    p2 = math.sin((12*x) - 4)
    return p1 * p2
```

```python
def g(x):
    return math.sin(10*x)
```

```python
def death(x, infinity=20):
    if (g(x) <= 0):
        return f(x)
    else:
        return infinity
```

```python
def static(x, r=20, b=1):
    return f(x) + s(x, r, b)
```

```python
def s(x, r=20, b=1):
    return r * (max(0, g(x))**b)
```

# Task 2

```python
x = numpy.linspace(0, 1, 1000)

y = []
for i in x:
    y.append(f(i))
pyplot.plot(x, y)

y = []
for i in x:
    y.append(g(i))
pyplot.plot(x, y)

y = []
for i in x:
    y.append(s(i))
pyplot.plot(x, y)

y = []
for i in x:
    y.append(death(i))
pyplot.plot(x, y)

y = []
for i in x:
    y.append(static(i))
pyplot.plot(x, y)

pyplot.axhline(0, color="black", ls="dotted")
pyplot.xlabel("x")
pyplot.ylabel("y")
pyplot.show()
```

# Task 3

```python
x = numpy.linspace(0, 1, 1000)
y = 1000000
yx = 0
for i in x:
    value = f(i)
    if (value < y):
        y = value
        yx = i
print((yx, y))
```

```python
def staticN(x, r=20, b=1):
    return f(x) - s(x, r, b)
```
