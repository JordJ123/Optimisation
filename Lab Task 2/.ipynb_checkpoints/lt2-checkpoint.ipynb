```python
import matplotlib.pyplot as pyplot
import numpy
```

# Task 2

```python
def squaredError(theta, X, Y):
    value = 0
    length = min(len(X), len(Y))
    for i in range(0, length):
        r = Y[i] - (theta['m'] * X[i]) - theta['c']
        value += r**2
    return value
theta = {'m':0.6014, 'c':2.1958}
print(squaredError(theta, [1, 2, 3, 4, 5], 
                   [2, 4, 5, 4, 5]))
```

# Task 3

```python
def gridSearch(k):
    current = 1000000
    for i in range(0,k):
        m = 0 + ((1/(k-1)) * i)
        for j in range(0,k):
            c = 2 + ((1/(k-1)) * j)
            theta = {'m':m, 'c':c}
            temp = squaredError(theta, [1, 2, 3, 4, 5], 
                [2, 4, 5, 4, 5])
            if temp < current:
                current = temp
    return current
print(gridSearch(100))
```

```python
def randomSearch(k):
    current = 1000000
    m = numpy.random.uniform(low=0, high=1, size=k)
    c = numpy.random.uniform(low=2, high=3, size=k)
    for i in range(0,k):
        theta = {'m':m[i], 'c':c[i]}
        temp = squaredError(theta, [1, 2, 3, 4, 5], 
            [2, 4, 5, 4, 5])
        if temp < current:
                current = temp
    return current
print(randomSearch(100))
```

```python
gs = []
rs = []
for i in range(0,100):
    gs.append(gridSearch(100))
    rs.append(randomSearch(100))
pyplot.boxplot(gs) #Deterministic
pyplot.show()
pyplot.boxplot(rs) #Stochastic
pyplot.show()
```
